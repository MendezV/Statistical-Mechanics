```python
#carga los datos del indice S&P 500 y los grafica con respecto a días 
# de intercambio de acciones desde el 15 de Octubre de 1982

data= loadtxt("./SandPConstantDollars.dat")
t=data[:,0]
SP=data[:,1]
N=len(SP)
plot(t,SP)
```

```python
#función que devuelve el cambio porcentual del índice S&P 500
#en un intervalo de tiempo definido como lag=1 para dias lag=5 para semanas
# y lag = 262 para años

def P(lag):
    ratios = SP[lag:N:lag]/SP[0:N-lag:lag]
    P=100*(ratios-1)
    return P
    
```

```python
def HistLag(lag):
    Daily = P(lag)
    (n,bins,patches) = pylab.hist(Daily,bins=100)
    print Daily
```

```python
HistLag(1) #histograma de las variaciones diarias en el mercado de valores de acuerdo a S&P 500
```

```python
HistLag(5) #histograma de las variaciones semanales en el mercado de valores de acuerdo a S&P 500
```

```python
HistLag(262)  #histograma de las variaciones anuales en el mercado de valores de acuerdo a S&P 500
```
